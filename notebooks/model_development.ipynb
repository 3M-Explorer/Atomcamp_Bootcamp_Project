# LinearRegression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors # import the colors module for normalization

# Assuming df is your DataFrame and the necessary columns are already present and processed

# Define features and target variable
X = df[['Synthesis technique','Catalyst loading (mg/cm2)','Electrode/Substrate','Morphology', 'Potential']]
y = df['FE (CO)']  # Target column

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)

# Make predictions on the training and test data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Evaluate the model for training and testing data
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

# Define a colormap
cmap = plt.get_cmap('viridis')

# Create a normalization object - adjust vmin and vmax to your data range if needed
norm = mcolors.Normalize(vmin=y_test_pred.min(), vmax=y_test_pred.max())

# Scatter plot with square markers and gradient effect
scatter = plt.scatter(y_test, y_test_pred, c=y_test_pred, cmap=cmap, marker='s', s=100, alpha=0.7, edgecolors='k', norm=norm)

# Plot the 45-degree reference line (Actual = Predicted)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)

# Title and axis labels
plt.title(f'LinearRegression: Actual vs Predicted\n'
          f'Training RÂ²: {r2_train:.2f}, Testing RÂ²: {r2_test:.2f}\n'
          f'Training MSE: {mse_train:.2f}, Testing MSE: {mse_test:.2f}', fontsize=14)
plt.xlabel('Actual Values of FE', fontsize=14)
plt.ylabel('Predicted Values of FE', fontsize=14)

# Grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Define features and target variable
X = df[['Synthesis technique','Catalyst loading (mg/cm2)','Electrode/Substrate','Morphology', 'Potential']]
y = df['FE (CO)']  # Target column

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

# Initialize the Ridge regression model
ridge = Ridge()

# Define the hyperparameters for GridSearchCV
param_grid = {
    'alpha': [0.1, 1, 10, 100, 1000]  # Regularization strength
}

# Initialize GridSearchCV to find the best alpha using cross-validation
grid_search = GridSearchCV(estimator=ridge, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best alpha
best_alpha = grid_search.best_params_['alpha']
print(f"Best alpha: {best_alpha}")

# Use the best alpha to fit the final model
best_ridge = Ridge(alpha=best_alpha)
best_ridge.fit(X_train, y_train)

# Make predictions with the tuned model
y_pred_best = best_ridge.predict(X_test)

# Evaluate the model
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

# Display the evaluation metrics for the best model
print(f'Mean Squared Error (Best): {mse_best:.2f}')
print(f'R-squared (Best): {r2_best:.2f}')
```

# Support Vector Machine

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Define features and target
X = df[['Synthesis technique', 'Catalyst loading (mg/cm2)', 'Electrode/Substrate', 'Morphology', 'Potential']]
y = df['FE (CO)']  # Target column

# Function to evaluate model performance for different train-test ratios
def evaluate_model(train_size):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1 - train_size, random_state=42)

    # Initialize and train the SVR model
    svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)  # Hyperparameters of SVR model
    svr_model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = svr_model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

# Define a wide range of train-test split ratios to evaluate (from 0.1 to 0.9)
train_sizes = np.arange(0.1, 1.0, 0.05)  # Range from 0.1 to 0.95 in increments of 0.05
results = {'Train Size': [], 'MSE': [], 'R-squared': []}

# Evaluate for different train-test ratios
for train_size in train_sizes:
    mse, r2 = evaluate_model(train_size)
    results['Train Size'].append(train_size)
    results['MSE'].append(mse)
    results['R-squared'].append(r2)

# Print the evaluation results
for i in range(len(train_sizes)):
    print(f"Train Size: {results['Train Size'][i]:.2f} - MSE: {results['MSE'][i]:.2f}, R-squared: {results['R-squared'][i]:.2f}")
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Define features and target
X = df[['Synthesis technique', 'Catalyst loading (mg/cm2)', 'Electrode/Substrate', 'Morphology', 'Potential']]
y = df['FE (CO)']  # Target column

# Split the data into training and testing sets with a train size of 0.35
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.65, random_state=42)

# Initialize the base SVR model (it will be used for grid search)
svm_model = SVR(C=1.0, epsilon=0.1)  # Base SVR model without kernel specified

# Define the parameter grid for kernel
param_grid_kernel = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']  # Different kernel types
}

# Perform Grid Search for kernel parameter
grid_search_kernel = GridSearchCV(svm_model, param_grid_kernel, scoring='neg_mean_squared_error', cv=5)
grid_search_kernel.fit(X_train, y_train)

# Get the best kernel and the best model
best_kernel = grid_search_kernel.best_params_
best_model_kernel = grid_search_kernel.best_estimator_

# Make predictions on both training and test sets using the best model
y_pred_train_kernel = best_model_kernel.predict(X_train)
y_pred_test_kernel = best_model_kernel.predict(X_test)

# Evaluate the model on both training and testing sets
mse_train_kernel = mean_squared_error(y_train, y_pred_train_kernel)
r2_train_kernel = r2_score(y_train, y_pred_train_kernel)
mse_test_kernel = mean_squared_error(y_test, y_pred_test_kernel)
r2_test_kernel = r2_score(y_test, y_pred_test_kernel)

# Display the evaluation metrics
print(f"Best Kernel: {best_kernel['kernel']}")
print(f"Training RÂ²: {r2_train_kernel:.2f}, Testing RÂ²: {r2_test_kernel:.2f}")
print(f"Training MSE: {mse_train_kernel:.2f}, Testing MSE: {mse_test_kernel:.2f}")


# Define colormap and normalization for the plot
cmap = plt.get_cmap('coolwarm')
norm = mcolors.Normalize(vmin=y_pred_test_kernel.min(), vmax=y_pred_test_kernel.max())

# Scatter plot with square markers and gradient color based on predictions
plt.figure(figsize=(5, 5))
scatter = plt.scatter(y_test, y_pred_test_kernel, c=y_pred_test_kernel, cmap=cmap, marker='s', s=100, alpha=0.7, edgecolors='k', norm=norm)

# Plot the 45-degree reference line (Actual = Predicted)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label="Ideal Fit")

# Title and axis labels
plt.title(f'SVM Regression: Actual vs Predicted\n'
          f'Training RÂ²: {r2_train_kernel:.2f}, Testing RÂ²: {r2_test_kernel:.2f}\n'
          f'Training MSE: {mse_train_kernel:.2f}, Testing MSE: {mse_test_kernel:.2f}', fontsize=14)
plt.xlabel('Actual Values of FE', fontsize=14)
plt.ylabel('Predicted Values of FE', fontsize=14)

# Grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Display the plot
plt.tight_layout()
plt.show()

```
